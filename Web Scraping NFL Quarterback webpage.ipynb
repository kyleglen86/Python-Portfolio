```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_starting_quarterbacks_in_the_National_Football_League'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
# print(soup.prettify())
```

```python
table = soup.find('table', class_= 'wikitable sortable')
```

```python
# Class of the table I need <table class="wikitable sortable jquery-tablesorter" style="text-align:left;"> 
```

```python
nfl_quarterback_titles = table.find_all('th')
```

```python
nfl_quarterback_titles
```

```python
nfl_table_titles = [title.text.strip() for title in nfl_quarterback_titles]
```

```python
print(nfl_table_titles)
```

```python
df = pd.DataFrame(columns = nfl_table_titles)

df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```
